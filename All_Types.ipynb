## _Ordinal Data Encoding_

#### 1. Label Encoder or Ordinal Encoder
* ___In label encoder you are going to give the unique labels to the unique values that present in the data.___ 
* ___You can use this to the output features only, you can't use this to input features.___ 
* ___Remainder use this for only Y values___
* ___It is not highly recommend for categorical data, because it choose the value based on alphabetically___
* ___You can use this for numerical categorical values, like 0 and 1.___

<img src=attachment:5a4d8e4b-3ff0-4084-a715-1157e66216d5.png height=400 width=400> 

#### 2. Target Guided Ordinal Encoding
* ___In this method we will calculate the mean of each categorical variable based on the output and rank them.___

<img src=attachment:76ee5579-30fa-494e-bcb0-d8f61f8c2399.png height=400 width=400> 

#### 3. Ordinal Encoder
* ___It's a opposite of label encoder, because label encoder used for target variables and it has disadvantage when we use categorical target variables___
* ___To overcome this, we use a Ordinal Encoder it is specifically prepared for categorical features, the value are ranging ( 0 to n )___

## _Nominal Data Encoding_
#### 1. One Hot Encoding
* ___It just create a new feature for how many unique feature do you have in the data.___
* ___Then it will add 1 when the feature is present in the row, otherwise it will add 0___

<img src=attachment:debb61e3-3e5e-4513-bfac-92b2ab831f0f.png height=400 width=400> 

#### 2. Binary Encoding
* ___Binary encoding is a combination of Hash encoding and one-hot encoding. In this encoding scheme, the categorical feature is first converted into numerical using an ordinal encoder. Then the numbers are transformed in the binary number. After that binary value is split into different columns.___

* ___Binary encoding works really well when there are a high number of categories. For example the cities in a country where a company supplies its products.___

<img src=attachment:b081c75e-51c2-41c7-bb14-847fcc26d5d9.png height=400 width=400> 

#### 3.Mean Encoding
* ___It is same like target guided ordinal encoding.___
* ___In this method, we will convert the categories into their mean values based on the output.___
* ___It is applicable for large number of categories that present in the single feature___

<img src=attachment:ac4c09db-786b-46d5-bee7-3468aa080bd3.png height=400 width=400>

# Target Guided Ordinal Encoding

* **Step1:**  First we need to convert our converting feature to string. 
* And take the first value of that. 
* **Step2:** And then groupby with target feature, and find the index. 
* **Step3:** Enumerate with zero. 
* **Step4:** Map with converting feature. 

```python
# step 1 
data_str=data['ord_0'].astype(str).str[0]
# step 2
data_index=data.groupby(['ord_0'])['target'].mean().sort_values().index 
# step 3 
dic={ k:i for i,k in enumerate(data_index, 0) } 
# step 4 
data['converting_feature']=data['ord_0'].map(dic)
```
